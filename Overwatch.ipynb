```python
#competitive season 1
comp_s1_win_pick_rate_all_rank = overwatch_comp_s1[overwatch_comp_s1["Skill Tier"]=="All"]
#drop extraneous columns
for col in comp_s1_win_pick_rate_all_rank.columns:
    columns = ["Hero", "Pick Rate, %", "Win Rate, %"]
    if col not in columns:
        #print(col)
        del comp_s1_win_pick_rate_all_rank[col]
comp_s1_win_pick_rate_all_rank.tail(50)
plt.figure(figsize=(20,20))
plt.scatter(comp_s1_win_pick_rate_all_rank["Pick Rate, %"], comp_s1_win_pick_rate_all_rank["Win Rate, %"], s=10)
for i in range(len(comp_s1_win_pick_rate_all_rank["Hero"])):
    plt.annotate(comp_s1_win_pick_rate_all_rank["Hero"].iloc[i], (comp_s1_win_pick_rate_all_rank["Pick Rate, %"].iloc[i], comp_s1_win_pick_rate_all_rank["Win Rate, %"].iloc[i] + 0.1), fontsize=10)
plt.xlabel("Pick Rate, %")
plt.ylabel("Win Rate, %")
plt.title("Season 1 All Rank")

s1_pearsonr, p_s1 = pearsonr(comp_s1_win_pick_rate_all_rank["Pick Rate, %"], comp_s1_win_pick_rate_all_rank["Win Rate, %"])
print(s1_pearsonr, p_s1)
```

Based on the Pearsonrs of s1 to s4 there seems to be a weak correlation forming which warrants additional investigation.

```python
#organise pick rate/win rate into one df
season_1_to_4_all_rank = pd.DataFrame()
new_cols = ["Hero", "s1_pickrate", "s1_winrate", "s2_pickrate", "s2_winrate", "s3_pickrate", "s3_winrate", "s4_pickrate", "s4_winrate",]
new_vals = [comp_s1_win_pick_rate_all_rank["Hero"], comp_s1_win_pick_rate_all_rank["Pick Rate, %"], comp_s1_win_pick_rate_all_rank["Win Rate, %"], comp_s2_win_pick_rate_all_rank["Pick Rate, %"], comp_s2_win_pick_rate_all_rank["Win Rate, %"], \
            comp_s3_win_pick_rate_all_rank["Pick Rate, %"], comp_s3_win_pick_rate_all_rank["Win Rate, %"], comp_s4_win_pick_rate_all_rank["Pick Rate, %"], comp_s4_win_pick_rate_all_rank["Win Rate, %"],]
#map new columns as keys and new values as values
col_val_mapping = dict(zip(new_cols, new_vals))
#unpack new column/new value pairs and assign them to the dataframe
season_1_to_4_all_rank = season_1_to_4_all_rank.assign(**col_val_mapping).reset_index(drop=True)
season_1_to_4_all_rank.head(20)
```

```python
#season_1_to_4_all_rank[season_1_to_4_all_rank.Hero == "Ana"].s1_pickrate
def season_1_to_4_all_rank_values(hero, i):
    pick_rate = list()
    win_rate = list()
    pick_rate += (season_1_to_4_all_rank[season_1_to_4_all_rank.Hero == hero].at[i, "s1_pickrate"], season_1_to_4_all_rank[season_1_to_4_all_rank.Hero == hero].at[i, "s2_pickrate"], \
                  season_1_to_4_all_rank[season_1_to_4_all_rank.Hero == hero].at[i, "s3_pickrate"], season_1_to_4_all_rank[season_1_to_4_all_rank.Hero == hero].at[i, "s4_pickrate"])
    win_rate += (season_1_to_4_all_rank[season_1_to_4_all_rank.Hero == hero].at[i, "s1_winrate"], season_1_to_4_all_rank[season_1_to_4_all_rank.Hero == hero].at[i, "s2_winrate"], \
                 season_1_to_4_all_rank[season_1_to_4_all_rank.Hero == hero].at[i, "s3_winrate"], season_1_to_4_all_rank[season_1_to_4_all_rank.Hero == hero].at[i, "s4_winrate"])
    return pick_rate, win_rate
plt.figure(figsize=(30,20))
for i in range(len(season_1_to_4_all_rank)):
    stats = season_1_to_4_all_rank_values(season_1_to_4_all_rank["Hero"][i], i)
    plt.plot(stats[0], stats[1], marker="o")
    plt.annotate(season_1_to_4_all_rank["Hero"][i], (stats[0][0], stats[1][0]))

plt.xlabel("Pick Rate, %")
plt.ylabel("Win Rate, %")
plt.title("Season 1-4 All Rank")
```
